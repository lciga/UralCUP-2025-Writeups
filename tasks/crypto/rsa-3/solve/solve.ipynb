```python
def bytes_to_long(s):
    ans = 0
    for i in s:
        ans <<= 8
        ans += ord(i)
    return ans

def long_to_bytes(n):
    ans = []
    while n > 0:
        ans.append(chr(mod(n, 256)))
        n >>= 8
    return "".join(ans[::-1])
```
